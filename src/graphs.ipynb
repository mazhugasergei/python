```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-20, 20, 100)
y = np.power(x, 2)

plt.title('Parabola: y = xÂ²')
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# x^2 - y^2 = -1 (vertical)
x1 = np.linspace(-5, 5, 500)
y1_pos = np.sqrt(1 + x1**2)
y1_neg = -np.sqrt(1 + x1**2)

# x^2 - y^2 = 1 (horizontal)
x2 = np.linspace(-5, -1, 500)
x3 = np.linspace(1, 5, 500)
y2_pos = np.sqrt(x2**2 - 1)
y2_neg = -np.sqrt(x2**2 - 1)
y3_pos = np.sqrt(x3**2 - 1)
y3_neg = -np.sqrt(x3**2 - 1)

plt.plot(x1, y1_pos)
plt.plot(x1, y1_neg)
plt.plot(x2, y2_pos)
plt.plot(x2, y2_neg)
plt.plot(x3, y3_pos)
plt.plot(x3, y3_neg)

plt.title("Hyperbolas: xÂ² - yÂ² = Â±1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.title("Sine curve: y = sin(x)")
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```
